```python
import pandas as pd
from pathlib import Path
```

```python
DATA_DIR = Path("california")

TARGET_MONTHS = [
    "202505",
    "202506",
    "202507",
    "202508",
    "202509",
    "202510",
    "202511",
    "202512",
    "202601",
    "202602",
    "202603",
    "202604",
    "202605",
]

files = sorted(
    [
        f
        for f in DATA_DIR.glob("CRMLSSold*.csv")
        if any(m in f.name for m in TARGET_MONTHS)
    ]
)

print(f"Found {len(files)} files:")
for f in files:
    print(f"  {f.name}")

dfs = []
for f in files:
    try:
        tmp = pd.read_csv(f, low_memory=False)
        tmp = tmp[
            (tmp["PropertyType"] == "Residential")
            & (tmp["PropertySubType"] == "SingleFamilyResidence")
        ].copy()
        tmp["source_file"] = f.name
        dfs.append(tmp)
        print(f"  {f.name}: {tmp.shape}")
    except Exception as e:
        print(f"  {f.name}: {e}")

df = pd.concat(dfs, ignore_index=True)
df["CloseDate"] = pd.to_datetime(df["CloseDate"])

print(f"\nTotal shape: {df.shape}")
print(f"Date range: {df['CloseDate'].min()} – {df['CloseDate'].max()}")
print(f"ClosePrice nulls: {df['ClosePrice'].isna().sum()}")

print("\nColumn count per file:")
for d, f in zip(dfs, files):
    print(f"  {f.name}: {d.shape[1]} cols")
```

```python
df.to_csv("california_sfr_202505_202605.csv", index=False)
print("Saved.")
```
